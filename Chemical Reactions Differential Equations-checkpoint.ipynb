```python
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def f(y,t):
    dydt= y * np.exp(-t)
    return dydt

t = np.linspace(0,6)
y0 = 1.0
y = odeint(f,y0,t)
print(y)

plt.plot(t,y)
plt.xlabel('Time(sec)')
plt.ylabel('y(t)')
```

```python
def rxn(C,t):
    Ca=C[0]
    Cb=C[1]
    
    k=2
    dAdt= -k * Ca
    dBdt=  k * Cb
    
    return(dAdt,dAdt)

print(rxn([0.1,0.5],0.5))

t = np.linspace(0,5,10)
C0 = [1,0]
C = odeint(rxn,C0,t)

plt.plot(t,C[:,0],'r--',linewidth=2.0)
plt.plot(t,C[:,1],'b-',linewidth=2.0)
plt.xlabel('Time (s)')
plt.ylabel('Concentration')
plt.legend(['Ca','Cb'])
```

```python
def rxn(Z,t):
    k1 = 1
    k2 = 1.5
    
    r1 = k1 * Z[0]*Z[1]
    r2 = k2 * Z[1]*Z[2]
    dAdt = -r1
    dBdt = -r1 -r2
    dCdt = r1 - r2
    dDdt = r2
    return(dAdt,dBdt,dCdt,dDdt)

t=np.arange(0,3.01,0.2)
print(t)
Z0 = [1,1,0,0]
Conc= odeint(rxn,Z0,t)

cA=Conc[:,0]
cB=Conc[:,1]
cC=Conc[:,2]
cD=Conc[:,3]
##Selectivity
S= np.empty(len(cC))
for i in range(len(cC)):
    if abs(cC[i]+cD[i]>1e-10):
        S[i] = cC[i] /cC[i] / (cC[i]+cD[i])
    else:
        S[i] = 1.0

plt.plot(t,cA)   
plt.plot(t,cB)
plt.plot(t,cC)
plt.plot(t,cD)
plt.plot(t,S)

plt.savefig('results.png')
#S= cC / (cc+cD)
```
